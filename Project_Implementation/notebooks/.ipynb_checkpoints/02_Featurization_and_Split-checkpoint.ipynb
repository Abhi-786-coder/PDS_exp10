# Phase 2: Multi-Representation Featurization & Scaffold Split
This notebook implements the Phase 2 pipeline from the project roadmap:
1. Load the cleaned Tox21 dataset from Phase 1
2. Perform a 4-way Murcko Scaffold Stratified Split (Train / Val / Calib / Test)
3. Run the fingerprint ablation study across 6 ECFP configurations
4. Extract and scale RDKit descriptors (train-only scaler)
5. Save all feature matrices and split indices for Phase 3/4

```python
import sys
sys.path.insert(0, '..')

import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import average_precision_score
from src.featurize import (
    FP_CONFIGS, build_fingerprint_matrix, build_descriptor_matrix, DescriptorScaler
)
from src.scaffold_split import scaffold_stratified_split
import warnings
warnings.filterwarnings('ignore')

print('✅ Libraries loaded.')
```

```python
TARGET_COLS = [
    'NR-AR', 'NR-AR-LBD', 'NR-AhR', 'NR-Aromatase', 'NR-ER', 'NR-ER-LBD',
    'NR-PPAR-gamma', 'SR-ARE', 'SR-ATAD5', 'SR-HSE', 'SR-MMP', 'SR-p53'
]

print('Loading cleaned dataset...')
df = pd.read_csv('../data/processed/tox21_cleaned.csv')
print(f'Loaded {len(df)} compounds with {len(TARGET_COLS)} endpoints.')
display(df.head())
```

## Step 1: 4-Way Murcko Scaffold Split
**Why scaffold split?** Random splitting causes structural leakage — similar molecules appear in both train and test, falsely inflating AUC. Murcko scaffold split ensures all four sets are structurally independent.

**Why 4-way?** The calibration split (10%) is required for Mondrian Conformal Prediction in Phase 5, which provides mathematically guaranteed uncertainty intervals under scaffold distribution shift.

```python
smiles_list = df['clean_smiles'].tolist()

train_idx, val_idx, calib_idx, test_idx = scaffold_stratified_split(
    smiles_list,
    train_size=0.70,
    val_size=0.10,
    calib_size=0.10,
    seed=42
)

# Verify splits are mutually exclusive
all_idx = np.concatenate([train_idx, val_idx, calib_idx, test_idx])
assert len(set(all_idx)) == len(df), 'ERROR: Overlapping indices detected!'
print('\n✅ All splits verified — mutually exclusive and complete.')
```

## Step 2: Fingerprint Ablation Study
We test 6 ECFP configurations (radius 2/3 × bits 1024/2048/4096) using a fast Logistic Regression baseline.
The winner is ranked by **Macro AUPRC** (not AUROC) — the correct metric for imbalanced data.
The best config will feed directly into ToxNet (Phase 4).

```python
smiles_train = [smiles_list[i] for i in train_idx]
smiles_val   = [smiles_list[i] for i in val_idx]
Y_train = df[TARGET_COLS].values[train_idx]
Y_val   = df[TARGET_COLS].values[val_idx]

ablation_results = []

for cfg in FP_CONFIGS:
    print(f"Testing {cfg['name']}...")
    X_train = build_fingerprint_matrix(smiles_train, radius=cfg['radius'], n_bits=cfg['n_bits'])
    X_val   = build_fingerprint_matrix(smiles_val,   radius=cfg['radius'], n_bits=cfg['n_bits'])

    auprc_scores = []
    for i, col in enumerate(TARGET_COLS):
        y_tr = Y_train[:, i]
        y_v  = Y_val[:, i]

        # Mask NaN labels
        train_mask = ~np.isnan(y_tr)
        val_mask   = ~np.isnan(y_v)

        if train_mask.sum() < 10 or val_mask.sum() < 5:
            continue
        if len(np.unique(y_tr[train_mask])) < 2:
            continue

        clf = LogisticRegression(max_iter=300, solver='lbfgs', class_weight='balanced')
        clf.fit(X_train[train_mask], y_tr[train_mask])
        y_prob = clf.predict_proba(X_val[val_mask])[:, 1]

        if len(np.unique(y_v[val_mask])) >= 2:
            auprc = average_precision_score(y_v[val_mask], y_prob)
            auprc_scores.append(auprc)

    macro_auprc = np.mean(auprc_scores) if auprc_scores else 0.0
    ablation_results.append({'Config': cfg['name'], 'Radius': cfg['radius'],
                              'Bits': cfg['n_bits'], 'Macro AUPRC': round(macro_auprc, 4)})
    print(f"  Macro AUPRC = {macro_auprc:.4f}")

ablation_df = pd.DataFrame(ablation_results).sort_values('Macro AUPRC', ascending=False)
print('\n--- Ablation Results ---')
display(ablation_df)
```

```python
best_cfg_name = ablation_df.iloc[0]['Config']
best_cfg = next(c for c in FP_CONFIGS if c['name'] == best_cfg_name)
print(f'✅ Best fingerprint config: {best_cfg_name} (Radius={best_cfg["radius"]}, Bits={best_cfg["n_bits"]})')
```

## Step 3: Build Full Feature Matrices Using Best Config
Now we build fingerprint matrices for all 4 splits using the winning config, and also build/scale RDKit descriptor matrices.

```python
smiles_calib = [smiles_list[i] for i in calib_idx]
smiles_test  = [smiles_list[i] for i in test_idx]

r, b = best_cfg['radius'], best_cfg['n_bits']

print(f'Building fingerprint matrices with {best_cfg_name}...')
X_fp_train = build_fingerprint_matrix(smiles_train, radius=r, n_bits=b)
X_fp_val   = build_fingerprint_matrix(smiles_val,   radius=r, n_bits=b)
X_fp_calib = build_fingerprint_matrix(smiles_calib, radius=r, n_bits=b)
X_fp_test  = build_fingerprint_matrix(smiles_test,  radius=r, n_bits=b)
print(f'  Train shape: {X_fp_train.shape}')

print('Building and scaling RDKit descriptor matrices...')
scaler = DescriptorScaler()
X_desc_train = scaler.fit_transform(build_descriptor_matrix(smiles_train))
X_desc_val   = scaler.transform(build_descriptor_matrix(smiles_val))
X_desc_calib = scaler.transform(build_descriptor_matrix(smiles_calib))
X_desc_test  = scaler.transform(build_descriptor_matrix(smiles_test))
print('✅ Feature matrices ready.')
```

## Step 4: Save All Artifacts for Phase 3/4

```python
import os, pickle
os.makedirs('../data/processed/splits', exist_ok=True)

# Save split indices
np.save('../data/processed/splits/train_idx.npy', train_idx)
np.save('../data/processed/splits/val_idx.npy',   val_idx)
np.save('../data/processed/splits/calib_idx.npy', calib_idx)
np.save('../data/processed/splits/test_idx.npy',  test_idx)

# Save feature matrices
np.save('../data/processed/splits/X_fp_train.npy', X_fp_train)
np.save('../data/processed/splits/X_fp_val.npy',   X_fp_val)
np.save('../data/processed/splits/X_fp_calib.npy', X_fp_calib)
np.save('../data/processed/splits/X_fp_test.npy',  X_fp_test)

np.save('../data/processed/splits/X_desc_train.npy', X_desc_train)
np.save('../data/processed/splits/X_desc_val.npy',   X_desc_val)
np.save('../data/processed/splits/X_desc_calib.npy', X_desc_calib)
np.save('../data/processed/splits/X_desc_test.npy',  X_desc_test)

# Save the scaler (needed for inference in Phase 6)
with open('../data/processed/splits/descriptor_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# Save ablation results
ablation_df.to_csv('../data/processed/splits/fp_ablation_results.csv', index=False)

print('✅ Phase 2 Complete. All feature matrices and indices saved to data/processed/splits/')
```
